# Utilizzando i dati riguardanti le partite calcolo il numero di vittorie e sconfitte nella stagione del 2023 per capire qual è la correlazione tra valore di mercato del club e la probabilità di successo in un match

```python
club_games_df = pd.read_csv(data_directory + 'club_games.csv')
games_df = pd.read_csv(data_directory + 'games.csv')
```

### Inserisco nel dataset club_games l'anno in cui si è tenuto ogni incontro

```python
merged_df = pd.merge(club_games_df, games_df[['game_id', 'season']], on='game_id', how='left')
merged_df
```

```python
club_games_df = merged_df[merged_df['season'] == 2023]
club_games_df
```

### Ora che ho la lista di partite aggiornata, conto il numero di vittorie e sconfitte per ogni club (un pareggio avendo is_win = 0 conterà come non vittoria)

```python
wins = club_games_df[club_games_df['is_win'] == True].groupby('club_id').size().reset_index(name='wins')
losses = club_games_df[club_games_df['is_win'] == False].groupby('club_id').size().reset_index(name='losses')

#ci sono partite di club che non hanno last_season = 2023 quindi le shapes di wins e losses sono maggiori di clubs_df
# wins.shape
# losses.shape
# clubs_df.shape
wins
```

### Inserisco i valori ottenuti in clubs_df e per le colonne che potrebbero avere NAN metto invece 0 (se non hanno mai registrato vittorie o sconfitte)

```python
clubs_df = pd.merge(clubs_df, wins, on='club_id', how='left')
clubs_df = pd.merge(clubs_df, losses, on='club_id', how='left')

clubs_df
```

```python
clubs_df['wins'] = clubs_df['wins'].fillna(0).astype(int)
clubs_df['losses'] = clubs_df['losses'].fillna(0).astype(int)

clubs_df
```

```python
clubs_df.columns
```

### Calcolo la correlazione tra total_market_value, vittorie e sconfitte

```python
correlation_df = clubs_df[['total_market_value', 'wins', 'losses']]
correlation_matrix = correlation_df.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix: Total Market Value, Wins, Losses')
plt.show()

# Analisi della correlazione per le singole variabili
sns.lmplot(x='total_market_value', y='wins', data=correlation_df, aspect=2, height=6)
plt.title('Correlation between Total Market Value and Wins')
plt.show()

sns.lmplot(x='total_market_value', y='losses', data=correlation_df, aspect=2, height=6)
plt.title('Correlation between Total Market Value and Losses')
plt.show()
```

### Come ci si aspetterebbe, nei tre grafici la correlazione tra vittorie aumenta con l'aumentare del valore di mercato e le vittorie diminuiscono con il diminuire del valore di mercato. Questo è indicato ad esempio nella heatmap da valori > 0 (correlazione positiva) per total_market_value e wins e valori < 0 (correlazione negativa) per total_market_value e losses (il valore = 0 indica assenza di correlazione).
### La linea di regressione nei due scatter plot quando ha inclinazione positiva indica correlazione positiva, negativa se ha inclinazione opposta.
### Nella zona in cui il valore dei club è basso, si hanno punti disposti in modo più disordinato, questo probabilmente perché nelle serie minori i risultati delle partite sono più imprevedibili.

# Il prossimo obiettivo è mettere in relazione la frequenza di vittorie con la percentuale di stranieri di ogni club per capire se avere tanti stranieri è un valore che incide sui successi

```python
clubs_df['foreigners_percentage'].describe()
```

### Ora raggruppo i club per la competizione di appartenenza in modo da vedere le percentuali di stranieri suddivise per campionato

```python
# Sostituisco i nomi delle competizioni per una migliore leggibilità
competition_mapping = {
    'ES1': 'La Liga',
    'FR1': 'Ligue 1',
    'GB1': 'Premier League',
    'IT1': 'Serie A',
    'L1': 'Bundesliga'
}

clubs_df.loc[:, 'domestic_competition_id'] = clubs_df['domestic_competition_id'].replace(competition_mapping)
clubs_df.groupby('domestic_competition_id')['foreigners_percentage'].describe()
```

I campionati con maggiori stranieri nei club sono quello inglese con la Premier League, quello italiano con la Serie A e quello francese con la Ligue 1

### Aggiungo il rate win di ogni club

```python
clubs_df.loc[:, 'total_games'] = clubs_df['wins'] + clubs_df['losses']
clubs_df.loc[:, 'win_rate'] = clubs_df['wins'] / clubs_df['total_games']

# Sostituisco i NAN (ad esempio generati per una divisione di 0) con 0
clubs_df.loc[:, 'win_rate'] = clubs_df['win_rate'].fillna(0)
clubs_df
```

### Mostro i risultati delle squadre italiane

```python
serie_a_clubs = clubs_df[clubs_df['domestic_competition_id'] == 'Serie A']

plt.figure(figsize=(12, 8))

indices = range(len(serie_a_clubs))

bar_width = 0.4

plt.bar(indices, serie_a_clubs['wins'], width=bar_width, label='Wins', color='b', align='center')
plt.bar(indices, serie_a_clubs['losses'], width=bar_width, label='Losses', color='r', align='edge')

plt.xlabel('Clubs')
plt.ylabel('Number of Games')
plt.title('Wins and Losses of Serie A Clubs')

plt.xticks(ticks=indices, labels=serie_a_clubs['name'], rotation=90)

plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

indices = range(len(serie_a_clubs))

plt.bar(indices, serie_a_clubs['win_rate'], color='b', align='center')

plt.xlabel('Clubs')
plt.ylabel('Win Rate')
plt.title('Win Rate of Serie A Clubs')

plt.xticks(ticks=indices, labels=serie_a_clubs['name'], rotation=90)

plt.tight_layout()
plt.show()
```

### Ora vediamo che percentuale di stranieri hanno le squadre e se questa è correlata al numero di vittorie

```python
serie_a_clubs[['win_rate', 'foreigners_percentage']].corr()
```

### La probabilità di vittoria dei club italiani sembrerebbe non correlata con il fatto che ci siano tanti stranieri nelle squadre. Questo credo perché in tutti i club italiani è presente un alto numero di stranieri. Controlliamo se anche chi vince poco ha tanti stranieri.

```python
win_rate_median = serie_a_clubs['win_rate'].median()

high_win_rate_clubs = serie_a_clubs[serie_a_clubs['win_rate'] > win_rate_median]
low_win_rate_clubs = serie_a_clubs[serie_a_clubs['win_rate'] <= win_rate_median]

plt.figure(figsize=(14, 7))

sns.kdeplot(high_win_rate_clubs['foreigners_percentage'], label='High Win Rate Clubs', fill=True)
sns.kdeplot(low_win_rate_clubs['foreigners_percentage'], label='Low Win Rate Clubs', fill=True)

plt.xlabel('Foreigners Percentage')
plt.ylabel('Density')
plt.title('Distribution of Foreigners Percentage by Win Rate Groups in Serie A')
plt.legend()

plt.show()
```

Dalla curva di distribuzione blu si evince che i club con maggiore win rate hanno una percentuale di stranieri concentrata intorno al 60%. La seconda curva marrone è più ampia e distribuita verso percentuali di stranieri maggiori; quindi nei club con basso win rate c'è alta variabilità nella percentuale di giocatori stranieri, con una leggera tendenza verso percentuali più alte.
Dal confronto si capisce che un'alta percentuale di giocatori stranieri non garantisce necessariamente un alto tasso di vittorie in Serie A, ma questo è dovuto al fatto che in circa tutti i club italiani c'è una buona percentuale di stranieri. 
Per questo la correlazione precedentemente calcolata era poco sotto allo 0

```python
filtered_clubs_df = clubs_df[['domestic_competition_id', 'foreigners_percentage', 'win_rate']]
correlations = filtered_clubs_df.groupby('domestic_competition_id').apply(lambda x: x['foreigners_percentage'].corr(x['win_rate']))

plt.figure(figsize=(12, 8))

correlations.plot(kind='bar', color='skyblue')

plt.xlabel('Domestic Competition ID')
plt.ylabel('Correlation between Foreigners Percentage and Win Rate')
plt.title('Correlation between Foreigners Percentage and Win Rate by Competition')

plt.tight_layout()
plt.show()
```

### A differenza del campionato italiano, quello tedesco sembra avere una correlazione maggiore tra il numero di stranieri e di vittorie per i club

```python
clubs_df[(clubs_df['domestic_competition_id'] == 'Bundesliga') & (clubs_df['win_rate'] > 0.5)]
```

Effettivamente il Bayer Leverkusen con un win rate di oltre 0.9 ha il 77.8% di stranieri

```python
clubs_df[(clubs_df['domestic_competition_id'] == 'Bundesliga') & (clubs_df['win_rate'] < 0.3)]
```

Allo stesso modo che ha un win rate molto basso ha una percentuale di stranieri generalmente sotto il 50%

```python
bundesliga_clubs = clubs_df[clubs_df['domestic_competition_id'] == 'Bundesliga']

plt.figure(figsize=(14, 7))

sns.kdeplot(data=bundesliga_clubs, x='win_rate', y='foreigners_percentage', cmap="Blues", fill=True)

plt.xlabel('Win Rate')
plt.ylabel('Foreigners Percentage')
plt.title('KDE Plot of Foreigners Percentage vs Win Rate in Bundesliga')

plt.show()
```

La concentrazione principale di densità si trova intorno al win rate di 0.25-0.5 e una percentuale di giocatori stranieri tra 30% e 65%.
La distribuzione della percentuale di giocatori stranieri è piuttosto ampia.
Non sembra esserci una forte correlazione visibile tra la percentuale di giocatori stranieri e il tasso di vittorie (altrimenti ci sarebbe una distribuzione dei dati che si concentra lungo una linea diagonale); entrambi i valori sono distribuiti in modo abbastanza indipendente.
Quello che potrebbe aver tirato su la correlazione rispetto gli altri campionati è il Bayer Leverkusen, precedentemente citato e qui rappresentato dalla macchia chiara in alto a destra rispetto a grafico.
