# Pulisco le tabelle togliendo le colonne che non mi servono

```python
clubs_df = clubs_df.drop(columns=['stadium_name', 'stadium_seats', 'coach_name', 'url'])
clubs_df.columns
```

```python
players_df = players_df.drop(columns=['agent_name', 'image_url', 'url', 'contract_expiration_date'])
players_df.columns
```

### Filtro i giocatori per l'ultima stagione (2023) e raggruppo per current_club_id in modo da poter calcolare la somma dei valori di mercato di un certo club e aggiornare la colonna del dataframe.

```python
filtered_players = players_df[players_df['last_season'] == 2023]

market_value_sum = filtered_players.groupby('current_club_id')['market_value_in_eur'].sum().reset_index()

market_value_sum.columns = ['club_id', 'total_market_value']

updated_clubs_df = pd.merge(clubs_df, market_value_sum, on='club_id', how='left')

updated_clubs_df['total_market_value'] = updated_clubs_df['total_market_value_y'].fillna(updated_clubs_df['total_market_value_x'])

updated_clubs_df = updated_clubs_df.drop(columns=['total_market_value_x', 'total_market_value_y'])

clubs_df = updated_clubs_df

clubs_df
```

# Utilizzando i dati riguardanti le partite calcolo il numero di vittorie e sconfitte nella stagione del 2023 per capire qual è la correlazione tra valore di mercato del club e la probabilità di successo in un match

```python
club_games_df = pd.read_csv(data_directory + 'club_games.csv')
games_df = pd.read_csv(data_directory + 'games.csv')
```

### Inserisco nel dataset club_games l'anno in cui si è tenuto ogni incontro

```python
merged_df = pd.merge(club_games_df, games_df[['game_id', 'season']], on='game_id', how='left')
merged_df
```

```python
club_games_df = merged_df[merged_df['season'] == 2023]
club_games_df
```

### Ora che ho la lista di partite aggiornata, conto il numero di vittorie e sconfitte per ogni club (un pareggio avendo is_win = 0 conterà come non vittoria)

```python
wins = club_games_df[club_games_df['is_win'] == True].groupby('club_id').size().reset_index(name='wins')
losses = club_games_df[club_games_df['is_win'] == False].groupby('club_id').size().reset_index(name='losses')

#ci sono partite di club che non hanno last_season = 2023 quindi le shapes di wins e losses sono maggiori di clubs_df
# wins.shape
# losses.shape
# clubs_df.shape
wins
```

### Inserisco i valori ottenuti in clubs_df e per le colonne che potrebbero avere NAN metto invece 0 (se non hanno mai registrato vittorie o sconfitte)

```python
clubs_df = pd.merge(clubs_df, wins, on='club_id', how='left')
clubs_df = pd.merge(clubs_df, losses, on='club_id', how='left')

clubs_df
```

```python
clubs_df['wins'] = clubs_df['wins'].fillna(0).astype(int)
clubs_df['losses'] = clubs_df['losses'].fillna(0).astype(int)

clubs_df
```

```python
clubs_df.columns
```

### Calcolo la correlazione tra total_market_value, vittorie e sconfitte

```python
correlation_df = clubs_df[['total_market_value', 'wins', 'losses']]
correlation_matrix = correlation_df.corr()

# correlation_matrix

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix: Total Market Value, Wins, Losses')
plt.show()

# Analisi della correlazione per le singole variabili
sns.lmplot(x='total_market_value', y='wins', data=correlation_df, aspect=2, height=6)
plt.title('Correlation between Total Market Value and Wins')
plt.show()

sns.lmplot(x='total_market_value', y='losses', data=correlation_df, aspect=2, height=6)
plt.title('Correlation between Total Market Value and Losses')
plt.show()
```

### Come ci si aspetterebbe, nei tre grafici la correlazione tra vittorie aumenta con l'aumentare del valore di mercato e le vittorie diminuiscono con il diminuire del valore di mercato. Questo è indicato ad esempio nella heatmap da valori > 0 (correlazione positiva) per total_market_value e wins e valori < 0 (correlazione negativa) per total_market_value e losses (il valore = 0 indica assenza di correlazione)
